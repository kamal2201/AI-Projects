```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model="llama-3.1-8b-instant", api_key=os.getenv("GROQ_API_KEY"), temperature=0)

result = llm.invoke("Hello, how are you?")
```

```python
result.content
```

```python
from langchain.agents import create_agent
from langchain.tools import tool
```

```python
@tool
def dummy_tool(inputText: str) -> str:
    """A dummy tool for testing, just returns the input in uppercase."""
    return f"{inputText.upper()}"
```

```python
my_tools = [dummy_tool]
```

```python
my_agent = create_agent(model=llm, tools=my_tools)
```

```python
agent_response = my_agent.invoke({"messages": [{"role": "user", "content": "Hey, I would like you to convert 'hello world' to uppercase"}]})
```
