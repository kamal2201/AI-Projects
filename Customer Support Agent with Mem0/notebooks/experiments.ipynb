```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
if not os.getenv("GROQ_API_KEY"):
    raise ValueError("GROQ_API_KEY not found in environment variables")
```

```python
mem0_config = {
    "llm": {"provider": "groq", "config": {"model": "llama-3.3-70b-versatile", "api_key": os.getenv("GROQ_API_KEY"), "temperature": 0}},
    "embedder": {"provider": "huggingface", "config": {"model": "sentence-transformers/all-MiniLM-L6-v2"}},
    "vector_store": {"provider": "chroma", "config": {"path": ".data/mem0-chroma"}}
}
```

```python
from mem0 import Memory

m = Memory.from_config(mem0_config)
```

```python
messages = [
    {"role": "user", "content": "Hi, I'm Alex. I love basketball and gaming."},
    {"role": "assistant", "content": "Hey Alex! I'll remember your interests."}
]
m.add(messages, user_id="alex")
```

```python
m.get_all(filters={"user_id": "alex"})
```
